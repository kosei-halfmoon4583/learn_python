1.関数とは  
 関数とはいろいろな処理をまとめて一つにしたものです。なぜ関数があるのでしょうか？  
 料理で例えてみます。カレーのレシピを例にとると、そのレシピを料理ロボットに記憶してもらいます。  
 そしてまたカレーが食べたくなった時に、ボタンひとつでカレーが作れるとしたらどうでしょうか？  
 しかもその料理ロボットは自分も使えるし、家族も使えるとしたらとても便利ですね、その料理ロボットが関数なのです。    
 関数の便利なところを3つ挙げてみます。  
 1.1 同じものを二回書く必要がない  
 1.2 １行で使い回しが可能  
 1.3 他の人も使うことができる(他の人が関数の中身を知る必要がない)。  
  
2.関数の種類  
 関数には2種類あります。自分で作る関数とPythonがあらかじめ用意してくれている関数です。  
 Pythonが予め用意してくれている関数は、「組み込み関数」と言います。print()関数や、max()関数などがそうです。  

3.関数の定義、引数、戻り値  
 関数を作ることを「関数を定義する」と言います。  
 def function(引数):  
     n = 引数 * 10    

```python
# 4.引数なしの関数
def say_hello():
    print("Hello World!")

say_hello()

# 5.引数ありの関数
def say_hello(greeting):
    print(greeting)

say_hello("Hello World!")

# 6.関数を変数に代入
def say_hello(greeting):
    print(greeting)

hello = say_hello
hello("Good morning!")

# 7.複数の引数がある関数
def add(num01, num02):
    print(num01 + num02)

add(6, 2)

# 8.戻り値がある関数
def add(num01, num02):
    return (num01 + num02)

print(add(7,2))

def add2(num01, num02):
    return (num01 + num02)

add_result = add2(8, 2)
print(add_result)
```

```python
# 9.デフォルト引数

def show_number(a=1, b=2):
    print(a)
    print(b)

show_number()

show_number(3,4)

# 注意点1
# デフォルト値の有無が混在する場合は、、、
# デフォルト値を渡さない引数を優先的に記述すること
# ※以下のような“a=1"の引数のことを「キーワード引数」と言い、bのような引数のことを「位置引数」と呼びます。
def show_number2(b, a=1):
    print(a)
    print(b)

show_number2(4)

# 注意点２
# デフォルト引数は関数定義時に一度だけ評価される
x = 1
def show_number3(a=x):
    print(a)

x = 10
show_number3()
```

```python
# 10.キーワード引数
def show_number(a,b,c):
    print(a)
    print(b)
    print(c)
    
show_number(a=1, c=10, b=100)

# a,b,cのように関数を定義する時に指定する引数のことを“仮引数”、1,10,100のように関数を呼び出す時に渡す値のことを“実引数”と呼びます。
```

```python
# 11.可変長位置引数

# 可変長一引数は、複数の引数を一つのタプルとして受け取ります。
# 関数を定義する時の仮引数の前にアスタリスク:*をひとつ書くことで、可変長位置引数を定義できます。

def show_number(a,b,*args):
    print(a)
    print(b)
    print(args)
    
show_number(1,2,3,4,5)

# また関数の引数にタプルを渡す場合、タプルの前にアスタリスク:*を書くと、args内のタプルが展開されて渡されます。
def show_number(a,b,*args):
    print(a)
    print(b)
    print(*args)
    
show_number(1,2,3,4,5)

# また可変長位置引数はタプルとして受け取るので、要素を一つ一つfor文で取り出すことができます。
def show_number(a,b,*args):
    print(a)
    print(b)
    for i in args:
        print(i)
    
show_number(1,2,3,4,5)

# 可変長位置引数が使われている代表例に、print()や max()があります。
# Pythonの組み込み関数として定義されているので、あまり意識することがありませんが、
# 仮引数には可変長位置引数が定義されています。
```

```python
# 12.可変長キーワード引数

# 可変長キーワード引数は複数のキーワード指定された引数をひとつの辞書として受け取れます。
# 関数を定義する時の仮引数の前に二つのアスタリスク:*を書くことで可変長キーワード引数として定義できます。
def show_number(**kwargs):
    print(kwargs)
    
show_number(a=1, b=2, c=3)

# また関数の引数に辞書を渡す場合、キーワード引数の前にアスタリスク:* 二つを書くと、kwargs内の辞書が展開されて渡されます。
def show_number(**kwargs):
    print(kwargs)

d = {"a":1,"b":2,"c":3}
show_number(**d)

# 可変長キーワード引数が使われている代表例が dict()関数です。
```

```python
# 13.関数内関数

# 関数内に定義された関数のことを「関数内関数」と言います。
def outer_func(a, b):
    def inner_func(c, d):
        return c - d
    
    x = inner_func(2, 1)
    y = inner_func(a, b)
    print(x)
    print(y)
    
outer_func(3, 1)

# 関数内関数は、プログラム中の他の処理では使われないけれど、ある関数内で似たような処理を繰り返したいときなどに使われます。
# 関数内関数を定義された関数の外からは呼び出せません。
```

```python
# 14.クロージャー

# クロージャーは関数の外で宣言された変数・引数を保持(記憶)した関数のことです。
def circle_area(pi, radius):
    return pi * radius * radius

# 定義した関数(circle_area()を丸括弧なしで変数:areaに引き渡すと、変数:areaには関数(circle_area())のオフジェクトが代入されます。
# 関数に丸括弧()を付けると、その関数(プログラム)が呼び出されて定義されたプログラムが実行されます。丸括弧()を付けないと関数オブジェクトとして扱われます。
area = circle_area
print(area)

# 上記の関数オブジェクトに丸括弧()を付けると、その関数を呼び出すことができます。
print(area(3.14, 2))

# 上記を踏まえて「クロージャ」を説明します。
def circle_area(pi):
    def calculation(radius): #この関数では“calculation”がクロージャです。
        return pi * radius * radius
    return calculation

area1 = circle_area(3.14)
area2 = circle_area(3)

print(area1(2))
print(area2(2))
```

```python
# 15.デコレーター

# デコレーターは、関数を修飾する関数のことです。デコレーターを使うことで関数を書き換えることなく、その関数に処理を追加・変更することができます。

# デコレーターを使わない場合を考える
def add(a, b):
    print(a + b)
    
print("Start!")
add(1,2)
print("Finish!!")

print("Start!!")
add(3,4)
print("Finish!!")

# クロージャを使って、上記の処理の"Start!!"と"Finish!!"を表示する部分を別の関数として定義してみます。
def info(func):
    def wrapper(a,b):
        print("Start!!")
        func(a,b)
        print("Finish!!")
    return wrapper

def add(a,b):
    print(a + b)

add_result = info(add)
add_result(1,2)
add_result(3,4)

# デコレーターを使って上記の処理を行う
def info(func):
    def wrapper(a,b):
        print("Start!!")
        func(a,b)
        print("Finish!!")
    return wrapper

@info # デコレーターを定義する(関数:addを書き換えることなく、関数:addに処理を追加することができます。
def add(a,b):
    print(a + b)

# あとはそのままaddを呼び出せば、デコレーターで修飾された結果が表示されます。
add(1,2)
add(3,4)
```

```python
# 15.デコレーター(#2)

# また今回は関数:addの引数が二つだったので、関数:wrapperの引数も二つにしましたが、このままでは引数が三つある関数などには使えません。
# そこでどんな関数に対しても使えるようにwrapperの引数は可変長位置引数と可変長キーワード引数にしておくと良いです。
def info(func):
    def wrapper(*args, **kwargs):
        print("Start!!")
        func(*args, **kwargs)
        print("Finish!!")
    return wrapper

@info # デコレーターを定義する(関数:addを書き換えることなく、関数:addに処理を追加することができます。
def add(a,b):
    print(a + b)

@info
def add2(c,d,e):
    print(c + d + e)

add(1,2)
add2(3,4,5)
```
